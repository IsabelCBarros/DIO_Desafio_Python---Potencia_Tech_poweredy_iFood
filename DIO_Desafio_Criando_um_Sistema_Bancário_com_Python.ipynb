# Objetivo Geral: #

# Criar um sistema bancrio com as operaes: sacar, depositar e visualizar extrato.





# Fomos contratados por um grande banco para desenvolver o seu novo sistema. Esse banco deseja modernizar suas operaes e para isso escolheu a linguagem Python. Para a primeira verso do sistema devemos implementar apenas 3 operaes: depsito, saque e extrato.

# Operao de depsito

Deve ser possvel depositar valores positivos para a minha conta bancria. A v1 do projeto trabalha apenas com 1 usurio, dessa forma no precisamos nos preocupar em identificar qual  o nmero da agncia e conta bancria. Todos os depsitos devem ser armazenados em uma varivel e exibidos na operao de extrato.

# Operao de saque

O sistema deve permitir realizar 3 saques dirios com limite mximo de R$ 500,00 por saque. Caso o usurio no tenha saldo em conta, o sistema deve exibir uma mensagem informando que no ser possvel sacar o dinheiro por falta de saldo. Todos os saques devem ser armazenados em uma varivel e exibidos na operao de extrato.

# Operao de extrato

Essa operao deve listar todos os depsitos e saques realizados na conta. No fim da listagem deve ser exibido o saldo atual da conta. Se o extrato estiver em branco, exibir a mensagem: No foram realizadas movimentaes.

image.png


```python
# Variveis iniciais
saldo = 0
limite_saque_diario = 500
numero_saques = 0
LIMITE_SAQUES = 3
extrato = []
valor_total_depositado = 0
valor_total_sacado = 0
primeiro_deposito_realizado = False

# Funo para exibir o cabealho do extrato
def exibir_cabecalho_extrato():
    print("======== Extrato =======")

# Funo para exibir o rodap do extrato
def exibir_rodape_extrato():
    print("========================")

# Funo para exibir o extrato da conta
def exibir_extrato():
    print()
    exibir_cabecalho_extrato()
    if not extrato:
        print("No foram realizadas movimentaes.")
    else:
        for movimentacao in extrato:
            print(movimentacao)
    saldo_real = valor_total_depositado - valor_total_sacado
    print(f"Saldo atual: R$ {saldo_real:.2f}")
    exibir_rodape_extrato()
    print()

# Loop principal
while True:
    # Exibir opes do menu
    print("""
    [1] Depositar
    [2] Sacar
    [3] Extrato
    [0] Sair
    """)

    opcao = input("=> ")

    if opcao == "1":
        # Depositar
        valor_deposito = float(input("Digite o valor do depsito: "))
        if valor_deposito > 0:
            saldo += valor_deposito
            extrato.append(f"Depsito: + R$ {valor_deposito:.2f}")
            print(f"O valor depositado foi R$ {valor_deposito:.2f}")
            valor_total_depositado += valor_deposito
            primeiro_deposito_realizado = True
        else:
            print("Valor de depsito invlido. O valor deve ser positivo.")

    elif opcao == "2":
        # Sacar
        if not primeiro_deposito_realizado:
            print("Saldo insuficiente para realizar o saque. Realize o primeiro depsito.")
        else:
            if numero_saques < LIMITE_SAQUES:
                valor_saque = float(input("Digite o valor do saque: "))
                if valor_saque > 0:
                    if valor_saque <= saldo:
                        if valor_saque <= limite_saque_diario:
                            saldo -= valor_saque
                            numero_saques += 1
                            extrato.append(f"Saque: - R$ {valor_saque:.2f}")
                            print(f"O valor sacado foi R$ {valor_saque:.2f}")
                            valor_total_sacado += valor_saque
                        else:
                            print("Limite dirio de saques atingido. Saque mximo  de R$ 500,00.")
                    else:
                        print("Saldo insuficiente.")
                else:
                    print("Valor de saque invlido. O valor deve ser positivo.")
            else:
                print("Limite dirio de saques atingido. Voc j realizou 3 saques hoje.")

    elif opcao == "3":
        # Extrato
        exibir_extrato()

    elif opcao == "0":
        # Sair
        print("Encerrando o programa...")
        break

    else:
        print("Operao invlida, por favor, selecione novamente a opo desejada")
```
